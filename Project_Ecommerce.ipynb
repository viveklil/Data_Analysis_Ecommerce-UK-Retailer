##  2. Check for missing values in all columns and replace them with the appropriate metric (Mean/Median/Mode)

```python
df.head(2)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(df['CustomerID'])
```

```python
df['CustomerID'] = df['CustomerID'].fillna(df["CustomerID"].mean())
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

## 3. Remove duplicate rows 


```python
df = df.drop_duplicates()
```

```python
df.shape
```

## 4. Remove rows which have negative values in Quantity column

```python
df[df['Quantity'] < 0]
```

```python
df = df.drop(df[df['Quantity'] < 0].index)
```

```python
df[df['Quantity'] < 0]
```

```python
df.shape
```

##  5. Add the columns - Month, Day and Hour for the invoice

```python
df.sample(10)
```

```python
df['Month'] = df['InvoiceDate'].str.split('/').str[0]
df['Date'] = df['InvoiceDate'].str.split('/').str[1]
df['Year_Time'] = df['InvoiceDate'].str.split('/').str[2]
df['Year'] = df['Year_Time'].str.split(' ').str[0]
df['Time'] = df['Year_Time'].str.split(' ').str[1]
df['Hour'] = df['Time'].str.split(':').str[0]
```

```python
df.head()
```

##  6. How many orders made by the customers? 

```python
df.head(2)
```

```python
orders = df.groupby(['CustomerID','Country'],as_index=False)['InvoiceNo'].count()
```

```python
orders.head(5)
```

```python
plt.subplots(figsize=(15,6))
plt.plot(orders.CustomerID,orders.InvoiceNo)
plt.xlabel('Customers ID')
plt.ylabel('Number of Orders')
plt.title('Number of Orders for different Customers')
```

## 7. TOP 5 customers with higher number of orders

##  11. How many orders per day?

```python
df.head(2)
```

```python
import datetime as dt
```

```python
df['InvoiceDate'] = pd.to_datetime(df.InvoiceDate, format='%m/%d/%Y %H:%M')
```

```python
df.insert(loc=4, column='Day', value=(df.InvoiceDate.dt.dayofweek)+1)
```

```python
df.head(3)
```

```python
df.groupby('InvoiceDate')['Day'].value_counts().sort_index()
```

```python
bar1 = df.groupby('InvoiceNo')['Day'].unique().value_counts().sort_index().plot(kind='bar',figsize=(15,8))
bar1.set_xlabel('Day',fontsize=15)
bar1.set_ylabel('Number of Orders',fontsize=15)
bar1.set_title('Number of orders for different Days',fontsize=15)
bar1.set_xticklabels(('Mon','Tue','Wed','Thur','Fri','Sun'), rotation='horizontal', fontsize=15)
```

## 13. How many orders for each country?

```python
df.head(2)
```

```python
df['Country'].value_counts()
```

```python
Total_orders = df['Country'].value_counts().sort_values()
```

```python
del Total_orders['United Kingdom'] ## we Delete this row to reprasent our data.
```

```python
plt.subplots(figsize=(20,12))
Total_orders.plot(kind='barh', fontsize=12,color='g')
plt.xlabel('Money Spent (Dollar)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.title('Money Spent by different Countries', fontsize=12)
plt.show()
```

Drop United Kingdom because it have a most number of orders as comparison to other Countries and it affects the Bar Graph.
